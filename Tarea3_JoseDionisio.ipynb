```python
#CONTEO DE DATOS ATIPICOS

precio_minimo = 1000
precio_maximo = 5000000
Total_Atipicos = ((Datos['selling_price'] < precio_minimo) | (Datos['selling_price'] > precio_maximo)).sum()
print("El total de datos atipicos es:  {}".format(Total_Atipicos))
```

```python
# FILTRAR DATOS ATIPICOS

Datos=Datos[(Datos['selling_price']<precio_maximo)&(Datos['selling_price']>precio_minimo)]
```

```python
Total_Atipicos = ((Datos['selling_price'] < precio_minimo) | (Datos['selling_price'] > precio_maximo)).sum()
print("El total de datos atipicos es:  {}".format(Total_Atipicos))
```

```python
# 4. DIVIDIR EL DATASET EN TRAIN Y TEST PARA EVALUAR CORRECTAMENTE EL MODELO

#PORCENTAJE DE DATOS PARA ENTRENAMIENTO (TRAIN) = 75% Y PARA PRUEBA(TEST) 25%

Datos = Datos[['selling_price', 'km_driven', 'year']]
X = Datos.drop('selling_price',axis=1)
Y = Datos['selling_price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state= 0)
```

```python
print (X_train)
```

```python
# 5. ENTRENAR EL MODELO CONFIGURANDO LOS DIFERENTES HIPERPARAMETROS

modelo_regresion = LinearRegression()
modelo_regresion.fit (X_train , Y_train)
Prediccion = modelo_regresion.predict(X_test)
```

```python
# 5. 
print (Y_test)
```

```python
# 5. 
print (Prediccion)
```

```python

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, Prediccion)
print(f"Mean Squared Error: {mse:.2f}")
```
